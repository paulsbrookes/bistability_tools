```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
from scipy.optimize import curve_fit
import os
import mpld3
from qutip import *
from collections import OrderedDict

mpld3.enable_notebook()

%matplotlib inline

dtypes = {'eps': np.float64,
         'fd': np.float64,
         'qubit_state': np.int32,
         't_levels': np.int32,
         'c_levels': np.int32,
         'fc': np.float64,
         'Ec': np.float64,
         'Ej': np.float64,
         'g': np.float64,
         'kappa': np.float64,
         'gamma': np.float64,
         'gamma_phi': np.float64,
         'n_t': np.float64,
         'n_c': np.float64,
         'end_time': np.float64,
         'snapshots': np.int32,
         'group_folder': str}

def decay(t, ar, ai, br, bi, T):
    signal_r = ar - br*np.exp(-t/T)
    signal_i = ai - bi*np.exp(-t/T)
    signal = np.array([signal_r, signal_i])
    return signal

def decay_flat(t, ar, ai, br, bi, T):
    index = int(len(t)/2)
    return decay(t[0:index], ar, ai, br, bi, T).flatten()

def decay_gen(a_ss):
    def decay_fixed(t, A, T):
        alpha = a_ss + A*np.exp(-t/T)
        return alpha
    return decay_fixed
```

```python
collated_results_dict = OrderedDict()
collated_popt = None

results_save = True

walk = os.walk('spectrum_calc')
for content in walk:
    directory = content[0]
    if os.path.exists(directory+'/results.csv'):
        
        settings = pd.read_csv(directory+'/settings.csv')
        settings = settings.set_index('job_index')
        settings = settings.T
        settings = settings.astype(dtypes)

        results = pd.read_csv(directory+'/results.csv')
        results = results.set_index('job_index')
        measurement_names = results.columns
        
        n_coords = 14
        coords = settings.iloc[0,0:n_coords]
        packaged_coords = []
        for coord in coords:
            packaged_coords.append([coord])

        for i, measurement in enumerate(measurement_names[0:3]):
            
            value = results.iloc[0,i]
            packaged_value = value
            for i in range(n_coords):
                packaged_value = [packaged_value]
            value_xarray = xr.DataArray(packaged_value, coords=packaged_coords, dims=coords.index)
            
            if measurement in collated_results_dict:
                collated_results_dict[measurement] = collated_results[measurement].combine_first(value_xarray)
            else:
                collated_results_dict[measurement] = value_xarray
                
collated_results = xr.Dataset(collated_results_dict)
```

```python
fig, axes = plt.subplots(1,1)

legend = []

a_op_re = collated_results['a_op_re'].squeeze()
n_t_levels = a_op_re.shape[1]
n_c_levels = a_op_re.shape[2]
for i in range(n_t_levels):
    for j in range(n_c_levels):
        sweep = a_op_re[:,i,j].dropna('fd')
        if sweep.shape[0] != 0:
            legend.append(str(sweep.t_levels.values[()]) + 'x' + str(sweep.c_levels.values[()]))
            sweep.plot.line(marker='o', ax=axes)

axes.legend(legend)
```
